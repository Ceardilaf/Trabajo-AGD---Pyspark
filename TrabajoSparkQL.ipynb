```python
# Agrege acá el código para importar las librerias

# La libreria para "encontrar el sevicio" de Spark
import findspark
findspark.init()

# Librerias para "gestionar el servicio" de Spark
from pyspark import SparkConf, SparkContext
from pyspark.sql import SQLContext, SparkSession

# Creamos una aplicación Spark en el Servicio
# Tenga cuidado con las tildes o caracteres especiales en el nombre de la app
AppSpark = SparkConf().setAppName("Evaluacion iSofware")

# definimos un espacio o contexto para la App
ContextoSpark=SparkContext(conf=AppSpark)

# inicio una sesión en el espacio de la App
SesionSpark = SparkSession(ContextoSpark)

# inicio del espacio o contexto SQL
ContextoSql = SQLContext(sparkContext=ContextoSpark, sparkSession=SesionSpark)
```

```python
# 1. Cargue los datos en la carpeta datalake y luego del /datalake al HDFS (Hadoop File System)
# Recuerda usar ! para ejecutar el comando en el shell. 
# Tu código a continuación...

!hdfs dfs -copyFromLocal asistencia.csv /tmp/
!hdfs dfs -copyFromLocal evaluacion.csv /tmp/
!hdfs dfs -ls /tmp/
```

```python
# 2. Cree dos tablas SparkSQL y almacene el csv en su correspondiente tabla.
# Observación: tenga especial cuidado con los encabezados de los archivos CSV.
# Usted puede considerar cambiar los encabezados de los CSV originales
# Tu código a continuación...

asistencia = ContextoSql.read.load("/tmp/asistencia.csv",
                                   format="csv",
                                   sep=",",
                                   inferSchema="true",
                                   header="true")

evaluacion = ContextoSql.read.load("/tmp/evaluacion.csv",
                                   format="csv",
                                   sep=",",
                                   inferSchema="true",
                                   header="true")
```

```python
# Renombrar columnas de asistencia
new_names = ['c1', 'c2', 'c3', 'c4']
data_asistencia = asistencia.toDF(*new_names)
data_asistencia.createOrReplaceTempView('data_asistencia')
```

```python
# Renombrar columnas de evaluación
new_names = ['c' + str(i + 1) for i in range(20)]
data_evaluacion = evaluacion.toDF(*new_names)
data_evaluacion.createOrReplaceTempView('data_evaluacion')
```

```python
# Mostrar tablas 
ContextoSql.sql("SHOW TABLES").show()
```

```python
# 2. Consulte el listado total de estudiantes (correos electrónicos) del 
# curso de Ingeniería de Software, ordenados alfabéticamente
# Tu código a continuación...

ContextoSql.sql("""
SELECT DISTINCT c2 FROM data_asistencia ORDER BY c2
""").show()
```

```python
# 3. Consulte la cantidad de asistencias registradas por estudiante; además, la fecha y hora de la primera asistencia
# y la fecha y hora de la última asistencia
# Tu código a continuación...

ContextoSql.sql("""
SELECT 
    c2 AS Estudiante, 
    COUNT(c2) AS Asistencias, 
    MIN(c1) AS Primera_Asistencia, 
    MAX(c1) AS Ultima_Asistencia
FROM 
    data_asistencia 
GROUP BY 
    c2
""").show()
```

```python
# 4. Consulte el listado de estudiantes que asistieron a 2 presentaciones o menos (una).
# Tu código a continuación...

ContextoSql.sql("""
SELECT 
    Estudiante, Asistencias
FROM 
    (
     SELECT 
         c2 AS Estudiante,
         COUNT(c2) AS Asistencias
     FROM 
         data_asistencia
     GROUP BY
         c2
    )
WHERE
    Asistencias <= 2
""").show()
```

```python
# 5. Consulte el listado de estudiantes que no asistieron a ninguna presentación.
# Tu código a continuación...

ContextoSql.sql("""
SELECT DISTINCT
    c2 AS Estudiante
FROM 
    data_evaluacion
WHERE 
    c2 LIKE '%@%' AND
    c2 NOT IN (SELECT DISTINCT c2 FROM data_asistencia)
""").show()
```

```python
# 6. Consulte los integrantes por cada equipo al que pertenecen.
# Tu código a continuación...

ContextoSql.sql("""
SELECT 
    c2 AS Estudiante, c3 AS Equipo
FROM 
    data_asistencia 
GROUP BY 
    c2, c3
ORDER BY 
    c3
""").show()
```

```python
# 7. Consulte la cantidad de asistentes por presentación, sin considerar los asistentes que pertenecen 
# al equipo que realizó la presentación.
# Tu código a continuación...

ContextoSql.sql("""
SELECT 
    c4 AS Presentacion, count(c2) AS Numero_Asistentes 
FROM 
    data_asistencia 
WHERE 
    REPLACE(c3,'Pertenezco a: ','') != REPLACE(c4,'Expone: ','') 
GROUP BY 
    c4
""").show()
```

```python
# 8. Consutar cuáles integrantes evaluaron a su propio equipo. Estas evaluaciones no serán válidas, pues un
# integrante no puede evaluar a su propio equipo.
# Tu código a continuación...

ContextoSql.sql("""
SELECT 
    c2 AS Estudiante
FROM 
    data_asistencia 
WHERE 
    REPLACE(c3,'Pertenezco a: ','') = REPLACE(c4,'Expone: ','') 
ORDER BY 
    c2
""").show()
```

```python
# 9. Consultar la nota promedio por cada ítem (1 al 15), y la nota promedio total del cada equipo. Recuerde que no
# son válidas las evaluaciones realizadas por los miembros del mismo equipo.
# Tu código a continuación...

promedios = ContextoSql.sql("""
SELECT 
    Evaluacion.c3 AS Equipo, 
    ROUND(AVG(Evaluacion.c4),3) AS P1, 
    ROUND(AVG(Evaluacion.c5),3) AS P2, 
    ROUND(AVG(Evaluacion.c6),3) AS P3, 
    ROUND(AVG(Evaluacion.c7),3) AS P4, 
    ROUND(AVG(Evaluacion.c8),3) AS P5, 
    ROUND(AVG(Evaluacion.c9),3) AS P6, 
    ROUND(AVG(Evaluacion.c10),3) AS P7, 
    ROUND(AVG(Evaluacion.c11),3) AS P8, 
    ROUND(AVG(Evaluacion.c12),3) AS P9, 
    ROUND(AVG(Evaluacion.c13),3) AS P10, 
    ROUND(AVG(Evaluacion.c14),3) AS P11, 
    ROUND(AVG(Evaluacion.c15),3) AS P12, 
    ROUND(AVG(Evaluacion.c16),3) AS P13, 
    ROUND(AVG(Evaluacion.c17),3) AS P14, 
    ROUND(AVG(Evaluacion.c18),3) AS P15,
    ROUND((AVG(Evaluacion.c4) + AVG(Evaluacion.c5) + AVG(Evaluacion.c6) + 
    AVG(Evaluacion.c7) + AVG(Evaluacion.c8) + AVG(Evaluacion.c9) + 
    AVG(Evaluacion.c10) + AVG(Evaluacion.c11) + AVG(Evaluacion.c12) + 
    AVG(Evaluacion.c13) + AVG(Evaluacion.c14) + AVG(Evaluacion.c15) + 
    AVG(Evaluacion.c16)+ AVG(Evaluacion.c17)+ AVG(Evaluacion.c18))/15,3) AS PT 
FROM 
    data_evaluacion AS Evaluacion
LEFT JOIN 
    data_asistencia AS Asistencia
ON 
    Evaluacion.c2 = Asistencia.c2
WHERE 
    REPLACE(Asistencia.c3,'Pertenezco a: ','') != REPLACE(Asistencia.c4,'Expone: ','')  
GROUP BY 
    Evaluacion.c3
""")

promedios.show()

promedios.createOrReplaceTempView('promedios')
```

```python
# 10. Consulte el mejor equipo evaluado por cada ítem (según nota promedio. En caso de empate mostrar todos los empatados) 
# y el mejor equipo según el promedio total.
# Tu código a continuación...

for i in range(15):
    print('Equipo con mejor promedio - Item #' + str(i + 1) + ':')
    ContextoSql.sql("SELECT Equipo, MAX(P"+ str(i + 1) +") AS Promedio FROM promedios GROUP BY Equipo ORDER BY Promedio DESC").show(1)

print('Equipo con mejor promedio total:')
ContextoSql.sql("SELECT Equipo, MAX(PT) AS Promedio FROM promedios GROUP BY Equipo ORDER BY Promedio DESC").show(1)
```

```python
# 11. Consulte el peor equipo evaluado por cada ítem (según nota promedio. En caso de empate mostrar todos los empatados) 
# y el peor equipo según el promedio total.
# Tu código a continuación...

for i in range(15):
    print('Equipo con peor promedio - Item #' + str(i + 1) + ':')
    ContextoSql.sql("SELECT Equipo, MAX(P"+ str(i + 1) +") AS Promedio FROM promedios GROUP BY Equipo ORDER BY Promedio ASC").show(1)

print('Equipo con peor promedio total:')
ContextoSql.sql("SELECT Equipo, MAX(PT) AS Promedio FROM promedios GROUP BY Equipo ORDER BY Promedio ASC").show(1)
```

```python
# 12. Consulte la lista de estudiantes con la correspondiente nota obtenida en la presentación 
# (nota promedio total de la evaluación obtenida por el equipo)
# Tu código a continuación...

ContextoSql.sql("""
SELECT 
    Evaluacion.c2 AS Estudiante, 
    ROUND((AVG(Evaluacion.c4) + AVG(Evaluacion.c5) + AVG(Evaluacion.c6) + 
    AVG(Evaluacion.c7) + AVG(Evaluacion.c8) + AVG(Evaluacion.c9) + 
    AVG(Evaluacion.c10) + AVG(Evaluacion.c11) + AVG(Evaluacion.c12) + 
    AVG(Evaluacion.c13) + AVG(Evaluacion.c14) + AVG(Evaluacion.c15) + 
    AVG(Evaluacion.c16) + AVG(Evaluacion.c17) + AVG(Evaluacion.c18))/15, 3) AS Promedio_Total 
FROM 
    data_evaluacion AS Evaluacion
LEFT JOIN 
    data_asistencia AS Asistencia
ON 
    Evaluacion.c2 = Asistencia.c2
WHERE 
    REPLACE(Asistencia.c3,'Pertenezco a: ','') != REPLACE(Asistencia.c4,'Expone: ','')  GROUP BY Evaluacion.c2 ORDER BY Estudiante
""").show()
```
